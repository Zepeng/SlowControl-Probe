```python
import pandas as pd
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt   
import numpy as np
import os
```

```python
def data_read(file_names,things_to_pop):
    temp=[]
    for i in range(len(file_names)):
        temp.append(pd.read_csv(file_names[i],header=0))
    for j in range(len(temp)):
        for l in range(len(things_to_pop)):
            things_to_pop[l].append([])
        columns_list = list(temp[j])
        for k in range(len(temp[j][columns_list[1]])):
            temp_1=temp[j][columns_list[0]][k]
            temp_2=float(temp[j][columns_list[1]][k])
            temp_3=float(temp[j][columns_list[2]][k])
            temp_4=float(temp[j][columns_list[3]][k])
            temp_5=k
            things_to_pop[0][j].append(temp_1)
            things_to_pop[1][j].append(temp_2)
            things_to_pop[2][j].append(temp_3)
            things_to_pop[3][j].append(temp_4)
            things_to_pop[4][j].append(temp_5)
            
def data_read_2(directory,file_names,things_to_pop):
    temp=[]
    for i in range(len(file_names)):
        if file_names[i].endswith(".csv"):
            temp.append(pd.read_csv(directory+file_names[i],header=0))
    for j in range(len(temp)):
        columns_list = list(temp[j])
        for k in range(len(temp[j][columns_list[1]])):
            temp_1=temp[j][columns_list[0]][k]
            temp_2=float(temp[j][columns_list[1]][k])
            temp_3=float(temp[j][columns_list[2]][k])
            temp_4=float(temp[j][columns_list[3]][k])
            things_to_pop[0].append(temp_1)
            things_to_pop[1].append(temp_2)
            things_to_pop[2].append(temp_3)
            things_to_pop[3].append(temp_4)
    things_to_pop[4].append(np.arange(0,len(things_to_pop[3])))
            

def get_tics(data,interval):
    tics=[[],[]]
    for i in range(len(data)):
        if i%interval==0:
            temp=data[i].split(" ")
            tics[0].append(i)
            tics[1].append(temp[1])
    return tics

def p_plot(x_data,y_data,title,lables,tics):
    plt.figure(figsize=(10, 5))
    for i in range(len(y_data)):
        plt.plot(x_data,y_data[i],label=lables[i])
    plt.ylim(21)
    plt.title(title)
    plt.xlabel("Time")
    plt.ylabel("Temperature (\u00b0C)")
    plt.xticks(tics[0],tics[1],rotation='vertical', fontsize=10)
    plt.legend(loc='best')
    plt.show()
```
